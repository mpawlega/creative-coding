### OpenCV Testing

```python
# imports
import cv2 as cv
import numpy as np
```

### Video Capture

```python
vc = cv.VideoCapture(0)  # select video source

if not vc.isOpened():
    print("Cannot open camera/video")
    exit()

while True:
    ret, frame = vc.read()  # ret = bool , frame = image array
    if not ret:
        print("Failed to grab frame")
        break

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)    # grayscale
    ret, thresh = cv.threshold(gray, 100, 255, cv.THRESH_BINARY)  # threshold
    contours, hierarchy = cv.findContours(thresh, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)


    #edges = cv.Canny(frame, 100, 20)
    final = cv.drawContours(thresh, contours, -1, (0, 255, 0), 1)

    cv.imshow("preview", final)  # display the frame
    key = cv.waitKey(30) & 0xFF  # wait ~30ms, check key
    if key == 27:  # ESC key to exit
        break

vc.release()
cv.destroyAllWindows()
```
